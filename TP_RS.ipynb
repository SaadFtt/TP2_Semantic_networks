**1. Importation des bibliothèques**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

**2. Creation du réseau sémantique**

```python
# Initialiser le graphe
G = nx.DiGraph()

# Ajouter des nœuds (concepts)
concepts = ["chien", "animal", "viande", "nourriture", "chat", "être vivant"]
G.add_nodes_from(concepts)

# Ajouter des relations (arêtes) avec des étiquettes
relations = [
    ("chien", "animal", "est un"),
    ("chien", "viande", "mange"),
    ("viande", "nourriture", "est un"),
    ("chat", "animal", "est un"),
    ("animal", "être vivant", "est un"),
    ("être vivant", "respirer", "a besoin de")
]

# Ajouter les arêtes au graphe avec des étiquettes
for source, target, relation in relations:
    G.add_edge(source, target, label=relation)
```

**3-Visualiser le réseau sémantique**

```python
# Positionnement des nœuds pour une disposition claire
pos = nx.spring_layout(G)

# Dessiner les nœuds et les arêtes
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue", font_size=10, font_weight="bold", arrowsize=20)

# Ajouter les étiquettes des relations (arêtes)
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color="red", font_size=9)

# Afficher le graphe
plt.title("Réseau Sémantique des Concepts et Relations")
plt.show()
```

**4-Affichage des relation**

```python
# Afficher les relations associées au concept "chien"
print("Relations associées au concept 'chien':")
for neighbor in G.neighbors("chien"):
    print(f"Chien {G['chien'][neighbor]['label']} {neighbor}")

# Vérifier et afficher la relation entre "chien" et "animal"
if G.has_edge("chien", "animal"):
    print("\nRelation entre 'chien' et 'animal':")
    print(f"Chien {G['chien']['animal']['label']} animal")
else:
    print("\nPas de relation directe entre 'chien' et 'animal'")
```
